```python
#plot a gaussoian distribution
import numpy as np
import matplotlib.pyplot as plt
import math

mu = 0
variance = 1
sigma = math.sqrt(variance)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
plt.plot(x, 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (x - mu)**2 / (2 * sigma**2) ), label='Gaussian Distribution')  
beta = 0.4
variance = 1*beta
sigma = math.sqrt(variance)
plt.plot(x, 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (x - mu)**2 / (2 * sigma**2) ), label='Gaussian Distribution with beta=0.4')
plt.title('Gaussian Distribution')
plt.xlabel('x')
plt.ylabel('Probability')
plt.legend()
plt.show()
```
